```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU, BatchNormalization, Reshape, Flatten
from tensorflow.keras.optimizers import Adam




#<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<start-the-game>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>



#                                                         ::  THE OCEAN  ::
#                                             :: ONE PERCEPTRON, ONE OCEAN, ONE SPIRIT ::
#                                                            ::  start  ::


#                                              THE Generative Adversarial Networks (GANs)

# Generator model
def build_generator(latent_dim, output_shape):
    model = Sequential()
    model.add(Dense(256, input_dim=latent_dim))
    model.add(LeakyReLU(alpha=0.2))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Dense(512))
    model.add(LeakyReLU(alpha=0.2))
    model.add(BatchNormalization(momentum=0.8))
    model.add(Dense(output_shape, activation='linear'))
    model.add(Reshape((output_shape, 1)))
    return model

# Discriminator model
def build_discriminator(input_shape):
    model = Sequential()
    model.add(Flatten(input_shape=input_shape))
    model.add(Dense(512))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dense(256))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dense(1, activation='linear'))
    return model

# GAN model
def build_gan(generator, discriminator):
    discriminator.trainable = False
    model = Sequential()
    model.add(generator)
    model.add(discriminator)
    return model





#                                                         ::  THE OCEAN  ::
#                                                :: ONE MODEL, ONE OCEAN, ONE SPIRIT ::
#                                                            ::  end  ::




#<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<end-the-game>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>



#                                            ::POTENTIAL AND EXAMPLE DATA::
#                                                  ::    START   :::


# THE OCEAN DATA
file_path = r'C:\Users\Subhadeep Maishal\Music\JUP\test_data\CBN_INPUT.csv'
df = pd.read_csv(file_path, parse_dates=['DATE'])
features = df[['AAO', 'AO', 'DMI', 'NAO', 'ONI', 'PDO', 'SOI', 'TNA', 'TSA', 'WHWP', 'WP']]
target = df['MEIv2']
features = (features - features.min()) / (features.max() - features.min())
np.random.seed(100)




#                                            ::POTENTIAL AND EXAMPLE DATA::
#                                                  ::    END   :::



#                                       :: BUILD, OPTIMISE, COMPILE, and DEMONSTRATE ::
#                                                     :: START ::
# Build and compile the discriminator
discriminator = build_discriminator((features.shape[1],))  # DONT PALY WITH learning rate AND beta
discriminator.compile(loss='binary_crossentropy', optimizer=Adam(0.0005, 0.06), metrics=['accuracy'])

# Build and compile the generator
latent_dim = 100
generator = build_generator(latent_dim, features.shape[1])  # DONT PALY WITH learning rate AND beta
generator.compile(loss='binary_crossentropy', optimizer=Adam(0.0005, 0.06))

# Build and compile the GAN
discriminator.trainable = False
gan = build_gan(generator, discriminator)
gan.compile(loss='binary_crossentropy', optimizer=Adam(0.0005, 0.06))  # DONT PALY WITH learning rate AND beta


#                                         :: BUILD, COMPILE, and DEMONSTRATE ::
#                                                     :: END ::

 

    
#                                               ::  POWER OF GENs  ::
#                                                  ::  START  ::
# Training GAN
epochs = 100   #POWER OF EPOCHS
batch_size = 32  #BATCH IITKGP 2024


half_batch = int(batch_size / 2)

for epoch in range(epochs):
    idx = np.random.randint(0, features.shape[0], half_batch)
    real_data = features.iloc[idx].values  # Convert to numpy array

    noise = np.random.normal(0, 1, (half_batch, latent_dim))
    generated_data = generator.predict(noise)

    real_labels = np.ones((half_batch, 1))
    fake_labels = np.zeros((half_batch, 1))

    d_loss_real = discriminator.train_on_batch(real_data, real_labels)
    d_loss_fake = discriminator.train_on_batch(generated_data, fake_labels)
    d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

    noise = np.random.normal(0, 1, (batch_size, latent_dim))
    valid_labels = np.ones((batch_size, 1))

    g_loss = gan.train_on_batch(noise, valid_labels)

    if epoch % 100 == 0:
        print(f"{epoch} [D loss: {d_loss[0]} | D accuracy: {100 * d_loss[1]}] [G loss: {g_loss}]")
        
#                                                 ::  POWER OF GENs  ::
#                                                     ::  END  ::


        
        

        
        
#                                                  ::  PREDICT OCEAN  ::

# Generate synthetic output
num_samples = 420 #its a example
noise = np.random.normal(0, 1, (num_samples, latent_dim))
generated_data = generator.predict(noise)

#                                                  ::  PREDICTED OCEAN  ::





#                                                    ::  ONLY COLOR ::
#                                                     :: ONLY BLUE ::

# Plot actual and generated data
plt.figure(figsize=(12, 8))
plt.plot(df['DATE'], target, label='Actual', marker='o', color='black')
plt.plot(df['DATE'].iloc[-num_samples:], generated_data[:, 0], label='Generated', marker='o', color='blue')
plt.xlabel('Date', fontsize=14)
plt.ylabel('ENSO Values', fontsize=14)
plt.title('Actual vs Predicted ENSO Values Over Time', fontsize=16)
plt.legend(fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
